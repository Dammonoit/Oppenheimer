```python
#convert byte like strings into strings.
tweets_txt_new=[]
for tweet in tweets_txt:
    tweets_txt_new.append(tweet.decode('utf-8'))
tweets_txt_new
```

```python
import re
#cleaning the tweets.
clean_tweets_txt=[]
for tweet in tweets_txt_new:
    clean_tweets_txt.append(' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split()))
clean_tweets_txt
```

```python
#now let's calculate the sentiments.
from textblob import TextBlob
Positive_Tweets=[]
Negative_Tweets=[]
Neutral_Tweets=[]
Positive_Sentiment_Count=0
Negative_Sentiment_Count=0
Neutral_Sentiment_Count=0
for tweet in clean_tweets_txt:
    analysis = TextBlob(tweet)
    if analysis.sentiment.polarity > 0:
        Positive_Sentiment_Count+=1
        Positive_Tweets.append(tweet)
    elif analysis.sentiment.polarity == 0:
        Neutral_Sentiment_Count+=1
        Neutral_Tweets.append(tweet)
    else:
        Negative_Sentiment_Count+=1
        Negative_Tweets.append(tweet)
         
#so let's filter things out.
tl=len(clean_tweets_txt)
Positive_Sentiment=Positive_Sentiment_Count*100/tl
Negative_Sentiment=Negative_Sentiment_Count*100/tl
Neutral_Sentiment=Neutral_Sentiment_Count*100/tl
print("The Positive Sentiment of the Twiteratti's is {} out of {} Which is about {} %".format(Positive_Sentiment_Count,tl,Positive_Sentiment))
print("The Negative Sentiment of the Twiteratti's is {} out of {} Which is about {} %".format(Negative_Sentiment_Count,tl,Negative_Sentiment))
print("The Neutral Sentiment of the Twiteratti's is {} out of {} Which is about {} %".format(Neutral_Sentiment_Count,tl,Neutral_Sentiment))
print('#'*90)
print("The Positive Tweets are: \n",Positive_Tweets);print('#'*90)
print("The Negative Tweets are: \n",Negative_Tweets);print('#'*90)
print("The Neutral Tweets are: \n",Neutral_Tweets);print('#'*90)
```

```python
import matplotlib.pyplot as plt
 
# Data to plot
plt.figure(figsize=(16,10))
labels = 'Positive', 'Negative', 'Neutral'
sizes = [Positive_Sentiment_Count,Negative_Sentiment_Count,Neutral_Sentiment_Count]
colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.1, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%2.1f%%', shadow=True, startangle=90)
 
plt.axis('equal')
plt.savefig('Output_Plot/Sentiment_PIE.png', dpi=100)
plt.show()
```
